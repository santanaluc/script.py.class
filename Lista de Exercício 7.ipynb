<h3>2.	Dado uma matriz de ordem 3x3 faça um algoritmo que:
<p>a)	Preencher a matriz com números aleatórios na faixa entre 1 e 9 (inclusive)</p>
<p>b)	Calcule a soma dos elementos da primeira coluna;</p>
<p>c)	Calcule o produto dos elementos da primeira linha;</p>
<p>d)	Calcule a soma de todos os elementos da matriz;</p>
<p>e)	Calcule a soma do diagonal principal;</p>
</h3>

```python
import random

lista=[]
somaColuna = somaTotal = somaDiagonal = 0
produto = 1

for i in range(3):
    linha=[]
    for j in range(3):
        n=random.randint(1,9)
        linha.append(n)
    lista.append(linha)

for r in range(len(lista)):
    for c in range(len(linha)):
        print(lista[r][c], end=" ")
    print()

for s in range(len(lista)):
    somaColuna += lista[s][0]
    produto *= lista[0][s]
    somaDiagonal += lista[s][s]
    for t in range(len(linha)):
        somaTotal += lista[s][t]

print("-"*30)
print("A soma dos elementos da primeira Coluna é: ", somaColuna)
print("O produto dos elementos da primeira linha é: ", produto)
print("A soma de todos os elementos da matriz é: ", somaTotal)
print("A soma da diagonal principal é: ", somaDiagonal)
```

<h3>3.	Dado uma matriz de ordem 5x5 faça um algoritmo que verifique se a matriz é simétrica (aij=aji).</h3>

```python
ordem = 5
lista=[]
falso = 0

for i in range(ordem):
    linha=[]
    for j in range(ordem):
        n = int(input("Digite o número [{}],[{}]: ".format(i,j)))
        linha.append(n)
    lista.append(linha)

print("-"*30)

for r in range(len(lista)):
    for c in range(len(linha)):
        print(lista[r][c], end=" ")
        if(lista[r][c]!=lista[c][r]):
            falso +=1
    print()
    
print("-"*30)

if(falso>=1):
    print("A matriz não é Simétrica")
elif(falso==0):
    print("A matriz é simétrica")
```

<h3>5.	Considerando duas matrizes A e B de ordem  3 (3x3)  faca um algoritmo que some as duas matrizes  gerando a matriz  C. 
<p>a.	Os elementos da matriz C são a soma dos respectivos elementos de A e B.</p>
<p>b.	Onde as matrizes A e B deverão ser preenchidas com valores aleatórios na faixa de 10 a 20 e de 30 a 50, respectivamente.</p>
<p>c.	Exibir as matrizes A, B e C.</p>
</h3>

```python
import random
ordem=3
listaA=[]
listaB=[]
listaC=[]
for i in range(ordem):
    linhaA=[]
    linhaB=[]
    for j in range(ordem):
        r1=random.randint(10,20)
        r2=random.randint(30,50)
        linhaA.append(r1)
        linhaB.append(r2)
    listaA.append(linhaA)
    listaB.append(linhaB)

print()
print("Matriz A")
print("-"*30)

for r in range(len(listaA)):
    for c in range(len(linhaA)):
        print(listaA[r][c], end=" ")
    print()

print()
print("Matriz B")
print("-"*30)

for r in range(len(listaB)):
    for c in range(len(linhaB)):
        print(listaB[r][c], end=" ")
    print()

print()
print("Matriz C")
print("-"*30)

for i in range(len(listaA)):
    linhaC=[]
    for j in range(len(linhaA)):
        lc = listaA[i][j] + listaB[i][j]
        linhaC.append(lc)
    listaC.append(linhaC)

for r in range(len(listaC)):
    for c in range(len(linhaC)):
        print(listaC[r][c], end=" ")
    print()
```

<h3>8.	Dado uma matriz M , 5x5 faça um algoritmo que:
<p>a.	Preencha a matriz com números aleatórios na faixa de 1 a 25</p>
<p>b.	Calcule a soma dos elementos da primeira coluna;</p>
<p>c.	Calcule o produto dos elementos da primeira linha;</p>
<p>d.	Calcule a soma de todos os elementos da matriz;</p>
<p>e.	Calcule a soma do diagonal principal;</p>
<p>f.	Crie um menu para exibir a matriz e os resultados obtidos nos itens b,c,d,e.</p>
</h3>

```python
import random

def somaColuna(lista):
    somaC=0
    for i in range(len(lista)):
        somaC += lista[i][0]
    return somaC

def produtoLinha(lista):
    produtoL = 1
    for i in range(len(lista)):
        produtoL *= lista[0][i]
    return produtoL

def somaTotal(lista, linha):
    somaT = 0
    for i in range(len(lista)):
        for j in range(len(linha)):
            somaT += lista[i][j]
    return somaT

def somaDiagonal(lista):
    somaD = 0
    for i in range(len(lista)):
        somaD += lista[i][i]
    return somaD

ordem = 5
lista=[]
escolha = 0

for i in range(ordem):
    linha=[]
    for j in range(ordem):
        n=random.randint(1,25)
        linha.append(n)
    lista.append(linha)

print("Matriz A ")
print("-"*30)

for r in range(len(lista)):
    for c in range(len(linha)):
        print('{:^3}'.format(lista[r][c]), end=" ")
    print()

while(escolha!=5):
    print("-"*30)
    print("1. Soma da primeira coluna")
    print("2. Produto da primeira linha")
    print("3. Soma de todos os elementos")
    print("4. Soma da diagonal principal")
    print("5. Sair")
    escolha=int(input("Digite a opção desejada: "))
    print("-"*30)

    if(escolha==1):
        print("A soma da primeira coluna é: ",somaColuna(lista))
    elif(escolha==2):
        print("O produto da primeira linha é: ", produtoLinha(lista))
    elif(escolha==3):
        print("A soma de todos os elementos é: ", somaTotal(lista,linha))
    elif(escolha==4):
        print("A soma da diagonal principal é: ", somaDiagonal(lista))
    elif(escolha==5):
        print("Finalizando...")
        break
    else:
        print("Número invalido")
```
